# Recurrent Neural Network

## What is an Long Short-Term Memory (LSTM)

Long Short-Term Memory (LSTM) is a more advanced form of Hochreiter and Schmidhuber's Recurrent Neural Network (RNN) (GeeksForGeeks, 2025). LSTMs can detect long-term dependencies in sequential data, making them excellent for tasks such as language translation, speech recognition, and time series forecasting (GeeksForGeeks, 2025).

 Unlike typical RNNs, which use a single hidden state that is passed through time, LSTMs include a memory cell that stores information for lengthy periods of time, addressing the difficulty of learning long-term dependencies (GeeksForGeeks, 2025).

## Evalute and justify the dataset

This dataset consists of 25,671 text samples, each of which is a paragraph of writing with an associated author label. Five distinct authors are making up this dataset and the text samples associated with them are taken from popular works like Stephen King's 'The Shining' and Harmon and Roiland's 'Rick and Morty'. This structure makes it ideal for using a Long-Term Short-Term Memory Recurrent Neural Network (LSTM-RNN), which offers supervised classification using sequential text data paired with categorical labels (GeeksForGeeks, 2025). The paragraphs provide natural word sequences where syntax, order, and stylistic choices are important (GeeksForGeeks, 2025). This is in line with LSTM models' ability to capture long-term dependencies within languages (GeeksForGeeks, 2025).

The dataset demonstrates additionally qualities that make it suitable for LSTM and this classification problem (GeeksForGeeks, 2025). The first thing to address is the data quality, with more than 25000 paragraphs. This data consists of a large in volume of data which provide the LSTM with enough material to identify patterns as well as avoid issues such as underfitting that occurs when there is too little data available (Uma Mahesh, et al., 2025, p. 5). To ensure that every sequence can be directly mapped to its source, every paragraph is consistently associated with both an author and a filename.

Another quality that makes this dataset suitable for LSTM and this classification problem, is the formatting's consistency between samples. The preprocessing of data is easier as each entry is saved as a clear text paragraph free of unnecessary markup (symbols). This is crucial as the model's performance would be negatively affected by noisy or inconsistent data. However, there is some varied lengths of the paragraphs, meaning that some data samples are only a few words long, while others are many lines long (Sohail, 2024). However, the LSTM model can make use of techniques such as padding to accommodate varying sequence lengths, this variability is not a drawback but rather an opportunity that ensures both short and long paragraphs make a significant contribution to training (Sohail, 2024). Additionally, the sequential nature of the paragraph samples makes this dataset suitable for LSTM and this classification problem (GeeksForGeeks, 2025; Uma Mahesh, et al., 2025; p 1). The LSTM model would be able to identify complex stylistic elements and maintain context beyond word order, in contrast to more simpler models which reduce text to unordered collections of phrases (GeeksForGeeks, 2025; MATHLAB, 2025).

However, with these justifications some possible issues are still a possibility. Firstly, due to the fact that there is a class imbalance (over representation of one class) the model could develop biased and favour the majority class (Raghuwanshi, 2018). Additionally, with LSTM overfitting is a very big risk, particularly if the model learns to link a certain author with extremely uncommon words instead of capturing their overall style (Srivatsavaya, 2023). Even with these drawbacks, with proper preprocessing and techniques these drawbacks can be addressed or minimised. For example with the use of regularisation techniques such as early stopping, the model will be able to prevent overfitting, reduce training time as well as improve the generalisability of the model (GeeksForGeeks, 2025).

#### Basic text cleaning


````python
def clean_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove HTML tags (if any)
    text = re.sub(r'<[^>]+>', '', text)

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Remove or standardize punctuation - replace with space or remove
    text = re.sub(r'[^\w\s]', ' ', text)  # Replace punctuation with spaces

    text = re.sub(r'\d+', lambda x: num2words.num2words(int(x.group())), text)

    # Fix all contractions
    text = contractions.fix(text)

    # Remove extra whitespaces
    text = ' '.join(text.split())

    # Remove special characters and keep only letters and spaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Final whitespace cleanup
    text = ' '.join(text.split())

    return text

# Apply the function
df_cleaned['cleaned_paragraph'] = df_cleaned['paragraph'].apply(clean_text)
````

#### Lemmatisation

````python
# user1946217;
# 13 March 2013;
# wordnet lemmatization and pos tagging in python;
# 1;
# source code;
# Availabe at: <https://stackoverflow.com/questions/15586721/wordnet-lemmatization-and-pos-tagging-in-python>
# [Acccesed 26 June 2025].

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def lemmatize_text(text):
    words = word_tokenize(text)
    pos_tags = pos_tag(words)
    lemmatized_words = []
    for word, tag in pos_tags:
        pos = get_wordnet_pos(tag)
        lemma = lemmatizer.lemmatize(word, pos=pos)
        lemmatized_words.append(lemma)
    return ' '.join(lemmatized_words)

df_cleaned['cleaned_paragraph'] = df_cleaned['cleaned_paragraph'].apply(lemmatize_text)
````

````python
max_len = 70            # max length of each text
trunc_type = 'post'     # cut off extra words at the end
padding_type = 'post'   # pad with zeros at the end

# Convert text to sequences
train_sequences = tokenizer.texts_to_sequences(X_train)
test_sequences = tokenizer.texts_to_sequences(X_test)

# Pad them so they’re the same length
train_padded = pad_sequences(train_sequences, maxlen=max_len,
                             padding=padding_type, truncating=trunc_type)
test_padded = pad_sequences(test_sequences, maxlen=max_len,
                            padding=padding_type, truncating=trunc_type)
````

- This convert sentences into sequences of integers and ensure they all have the same length.
- The 'max_len' was set to **70** as based on the text length analysis done above most paragraphs are 90 words in length.

#### Building LSTM model

##### LSTM Architecture

````python
def build_model(hp):
    model = Sequential()

    # Embedding dimension
    embedding_dim = hp.Choice('embedding_dim', values=[32, 64, 128])
    model.add(Embedding(input_dim=vocab, output_dim=embedding_dim, input_length=max_len))

    # Dropout rate
    model.add(SpatialDropout1D(hp.Float('dropout_rate', min_value=0.1, max_value=0.5, step=0.1)))

    # LSTM units
    model.add(LSTM(hp.Int('lstm_units', min_value=32, max_value=128, step=32),
                   dropout=hp.Float('lstm_dropout', 0.1, 0.5, step=0.1),
                   recurrent_dropout=hp.Float('recurrent_dropout', 0.1, 0.5, step=0.1)))

    # Output layer
    model.add(Dense(len(y.unique()), activation='softmax'))

    # Optimizer with tunable learning rate
    lr = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
````

##### Hyperparameter Tuning

````python
## GeekForGeeks;
## 20 August 2025;
## Keras Tuner for Hyperparameter Optimization;
## 1;
## source code;
## Available at: https://www.geeksforgeeks.org/deep-learning/keras-tuner-for-hyperparameter-optimization;
## [Accessed on: 30 September 2025].

tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=1,
    directory='tuner_results',
    project_name='author_lstm'
)

tuner.search(
    train_padded, y_train,
    validation_split=0.2,
    epochs=5,
    batch_size=32,
    verbose=1
)
````

````python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"""
Best Hyperparameters:
- Embedding Dim: {best_hps.get('embedding_dim')}
- Dropout Rate: {best_hps.get('dropout_rate')}
- LSTM Units: {best_hps.get('lstm_units')}
- LSTM Dropout: {best_hps.get('lstm_dropout')}
- Recurrent Dropout: {best_hps.get('recurrent_dropout')}
- Learning Rate: {best_hps.get('learning_rate')}
""")
````

##### Train model

````python
best_model = tuner.hypermodel.build(best_hps)

history = best_model.fit(
    train_padded, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=32,
    callbacks=[EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)],
    verbose=1
)
````
